**Normal-Metal - Interface - Superconductor**

<div style="font-size:12px;">

A normal-metal–superconductor (NM–S) bilayer with a superconducting interfacial (I) region is considered. A current $I$ is applied, as schematically indicated in the figure.

<img src="./images/multilayer.png" style="display:block; margin-left:auto; margin-right:auto; width:60%;">

The following assumptions apply:

1. The interface does not act as an independent conductive channel.
2. The interface represents a bottleneck for current injection into the superconducting layer.
3. $T_{C}^{I} \gtrsim T_{C}^{S}$.


Under these assumptions, the equivalent resistance $R_{E}$ and the currents flowing through the circuit branches, $I_{NM}$ and $I_{I-S}$, for the three temperature cases illustrated are given by:

<table style="width:100%; text-align:center; background-color:white;">
<tr>
  <th><b>$T > T_{C}^{I}:$</b></th>
  <th><b>$T = T_{C}^{I}:$</b></th>
  <th><b>$T = T_{C}^{S}:$</b></th>
</tr>
<tr>
<td>

\begin{aligned}
R_{E}          &= \frac{R_{NM}(R_{I} + R_{S})}{R_{NM} + R_{I} + R_{S}} \\
I_{NM}         &= \frac{R_{E}}{R_{NM}} I \\
I_{I\text{-}S} &= \frac{R_{E}}{R_{I} + R_{S}} I
\end{aligned}

</td>
<td>

\begin{aligned}
R_E            &= \frac{R_{NM} R_S}{R_{NM} + R_S} \\
I_{NM}         &= \frac{R_{E}}{R_{NM}} I \\
I_{I\text{-}S} &= \frac{R_{E}}{R_{S}} I
\end{aligned}

</td>
<td>

\begin{aligned}
R_E            &= 0 \\
I_{NM}         &= 0 \\
I_{I\text{-}S} &= I
\end{aligned}

</td>
</tr>
</table>

Finally, the temperature dependence of the resistance of each component is modeled as:

\begin{aligned}
R_{NM}(T) &= R_{0-NM} \left( 1 + \alpha T \right), \\
R_{I}(T)  &= \frac{R_{0-I}}{2} \left[1 + \tanh\!\left(\frac{T - T_{C-I}}{\Delta T_{I}}\right)\right], \\
R_{S}(T)  &= \frac{R_{0-S}}{2} \left[1 + \tanh\!\left(\frac{T - T_{C-S}}{\Delta T_{S}}\right)\right].
\end{aligned}

Here, $T$ is the absolute temperature. For the normal-metal layer, $R_{0-NM}$ denotes the normal-state resistance and $\alpha$ is the linear temperature coefficient. For the interface and the superconductor, $R_{0-I}$ and $R_{0-S}$ are the normal-state resistances, $T_{C-I}$ and $T_{C-S}$ are the critical temperatures, and the superconducting transition widths are given by $\Delta T_{I}$ and $\Delta T_{S}.$
